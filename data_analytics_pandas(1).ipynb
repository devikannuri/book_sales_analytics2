<h1>First we need to Clean the data</h1>
<ul>
<li>We need to know any null values are there</li></ul>

```python
unclean_df.isnull().sum()
```

<li>There are null values in 3 columns</li> 

```python
unclean_df.rename(columns={'Publisher ':'Publisher'},inplace=True)
unclean_df.columns
```

```python
unclean_df.columns=unclean_df.columns.str.replace(' ','_')
unclean_df.columns
```

```python
unclean_df["Publishing_Year"].unique()
```

<ul>
    <li>Here we have some -ve year we replace those years with median of "Publishing_Year"</li>
</ul>

```python
publish=unclean_df["Publishing_Year"].median()
publish
unclean_df['Publishing_Year'] = unclean_df['Publishing_Year'].apply(lambda x: publish if x < 0 else x)
unclean_df
```

<ul>
<li>There are some null values in 3 columns</li>
</ul>

```python
unclean_df["Publishing_Year"].fillna(2013,inplace=True)
unclean_df["Book_Name"].fillna("UnKnown",inplace=True)
unclean_df["language_code"].fillna("UnKnown",inplace=True)
```

```python
unclean_df.info()
```

<ul>
    <li>
        Here "Publishing_Year" is in float value we need to convert them into integer
    </li>
</ul>

```python
unclean_df["Publishing_Year"]=unclean_df["Publishing_Year"].astype("int64")
```

```python
unclean_df
```

```python
unclean_df["genre"].unique()
```

<ul>
    <li>Here 'genre fiction' replace with 'genre'</li>
</ul>

```python
unclean_df["genre"]=unclean_df["genre"].replace("genre fiction","fiction")
```

```python
unclean_df["Publisher"].unique()
```

<ul>
    <li>Here 'HarperCollins Publishing' is replace with 'HarperCollins Publishers'</li>
</ul>

```python
unclean_df["Publisher"]=unclean_df["Publisher"].replace('HarperCollins Publishing','HarperCollins Publishers')
```

```python
clean_df=unclean_df
```

<h3>1.No_of_books_based on Author_rating</h3>

```python
author_ratings = noofbooks_by_author_rating["Author_Rating"]
count =  noofbooks_by_author_rating["count"]

# Plotting the pie chart

plt.figure(figsize=(5,5))
explode = (0.005, 0.005, 0.005, 0.005)
plt.pie(count, labels=author_ratings, autopct='%1.1f%%', startangle=140,explode=explode)
title_font = {'fontsize': 16, 'fontweight': 'bold', 'fontfamily': 'serif'}
plt.title('no.of.books_by_author_rating', fontdict=title_font)
plt.legend(title="Author Ratings", loc="right", labels=author_ratings,bbox_to_anchor=(1.5, 0.5))
plt.subplots_adjust(left=0.1, right=0.75)
plt.show()
```
